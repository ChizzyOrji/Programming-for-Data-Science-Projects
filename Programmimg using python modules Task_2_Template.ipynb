### Requirement FR11 - Investigate the effects of household income on levels of bias crimes

### Requirement FR12 - Test the hypothesis that household income has no effect on levels of bias crime

```python
#using pearson corrlation test

# import required library

from scipy.stats import pearsonr

# assign the data being analysed to variables
data1 = merged_data['median_household_income']
data2 = merged_data['avg_biascrimes_per_100k']

# pass the variables through pearson
correlation, p_value = pearsonr(data1, data2)

# print output to 4decimal places
print('correlation=%.4f, p_value=%.4f' % (correlation, p_value))
if p_value > 0.05:
    print('p_value is statistically insignificant')
else:
    print('p_value is statistically significant')
```

**H0:** There is no relationship between household income and average level of bias crimes

**H1:** There is a relationship between household income and average level of bias crimes

The closer the correlation value is to 1, the stronger the relationship between the two variables. From the analysis above,
the correlation value is 0.2958 which is closer to zero than it is to one. This means that the relationship between the variables is very weak and low. The variables might be a bit related to each other, but not so much.

The smaller the p-value, the stronger the evidence that the null hypothesis should and must be rejected.
When the p-value is less than 0.05, it is considered statistically significant and when it is higher than 0.05, it is not statistically significant and indicates a strong evidence for the null hypothesis. From the analysis, the p_value is 0.0435 and is less than 0.05, this means the relationship is statistically significant, hence the null hypothesis should be rejected. There is a slight relationship between household income and average level of bias crimes

### Requirement FR14 - Test the hypothesis that urbanisation has a significant effect on levels of bias crimes

```python
# import library

from scipy.stats import pearsonr

# assign the data being analysed to variables
data1 = merged_data['share_population_in_metro_areas']
data2 = merged_data['avg_biascrimes_per_100k']

# pass the variables through pearson
correlation, p_value = pearsonr(data1, data2)

# print output to 4decimal places
print('correlation=%.4f, p_value=%.4f' % (correlation, p_value))
if p_value > 0.05:
    print('p_value is statistically insignificant')
else:
    print('p_value is statistically significant')
```

**H0:** There is no relationship between urbanisation and average level of bias crimes

**H1:** There is a relationship between urbanisation and average level of bias crimes

The closer the correlation value is to 1, the stronger the relationship between the two variables. From the analysis above,
the correlation value is 0.2095 which is closer to zero than it is to one. This means that the relationship between the variables is very weak and low. The variables might be a bit related to each other, but not so much.

The smaller the p-value, the stronger the evidence that the null hypothesis should and must be rejected.
When the p-value is less than 0.05, it is considered statistically significant and when it is higher than 0.05, it is not statistically significant and indicates a strong evidence for the null hypothesis. From the analysis, the p_value is 0.1576 and this is higher than 0.05, hence the null hypothesis should not be rejected.

Therefore, there is no relationship between urbanisation and average level of bias crimes.
